# Create Database

With three csv files created, I can now create a new database.

```python
import duckdb

conn = duckdb.connect('final_db.db')

%load_ext sql
%sql conn --alias duckdb
%config SqlMagic.displaylimit = None
```

```python
%sql CREATE TABLE IF NOT EXISTS dim_customers AS SELECT * FROM 'datasets/customers_info.csv'
%sql CREATE TABLE IF NOT EXISTS fact_sales AS SELECT * FROM 'datasets/total_sales.csv'
%sql CREATE TABLE IF NOT EXISTS dim_products AS SELECT * FROM 'datasets/final_product.csv'
```

```python
%%sql

SELECT TABLE_NAME, COLUMN_NAME, DATA_TYPE
FROM INFORMATION_SCHEMA.COLUMNS
```
